```python
#Now let's read excel files
df = pd.read_excel('stock_data.xlsx')
```

```python
#So now if you want in the people column instead of n.a sam  walton should come
def convert_people_cell(cell):
    if cell == "n.a.":
        return "Sam Walton"
    return cell

df = pd.read_excel('stock_data.xlsx', 'Sheet1' , converters = {
    'people':convert_people_cell
})
```

```python
df
```

```python
#Now let's try this same logic in some other columns let's use generators as used above
def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell
df = pd.read_excel('stock_data.xlsx', 'Sheet1' , converters = {
    'eps': convert_eps_cell
})
```

```python
df
```

```python
#So now if you want to write this dataframe to a new excel file
#Also if you want to give a customized name to sheet1
df.to_excel('jammu_don.xlsx',sheet_name="Stocks")
```

```python
#If you want to start from a particular row and stop at a particular column
df.to_excel('jammu_don.xlsx',sheet_name = "Stocks" , startrow = 1 , startcol = 2)
```

```python
#If you don't want index then
df.to_excel('jammu_don.xlsx' , sheet_name = "Stocks" , index = False)
```

```python
#How can we add 2 dataframes in a particular excel file in different sheets
#For that we will use excel writer class available in module pandas
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})
df_weather = pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer , sheet_name = "Stocks")
    df_weather.to_excel(writer , sheet_name = "weather")
```
